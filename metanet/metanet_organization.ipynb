```python
import json
from pprint import pprint
import pandas as pd

with open("data/all_data.json") as f:
    data = json.load(f)
```

```python
ad_keys = ['Binding', 'CxnPattern', 'Entailment', 'Example', 'Frame', 'FrameFamily', 'Inference', 'LexicalUnit', 'Mapping', 'Metaphor', 'MetaphorFamily', 'RelationalConfiguration', 'Role']

ad_data = [data['Binding'], data['CxnPattern'], data['Entailment'], data['Example'], data['Frame'], data['FrameFamily'], data['Inference'], data['LexicalUnit'], data['Mapping'], data['Metaphor'], data['MetaphorFamily'], data['RelationalConfiguration'], data['Role']]

def make_df(dt):
    return pd.DataFrame(dt)

def make_df_T(dt):
    return pd.DataFrame(dt).T

ad_dfs = []

for i in ad_data:
    ad_dfs.append(make_df(i).T)
```

```python
for i in range(len(ad_dfs)):
    ad_dfs[i].reset_index().rename(columns={'index': 'name'}).to_csv(f'data/{ad_keys[i]}.csv', index=False)
```
